```python
import numpy as np
import pandas as pd
```

```python
column_names=['user_id','item_id','rating','timestamp']
df=pd.read_csv('u.data',sep='\t',names=column_names)
```

```python
df.head(5)
```

```python
movie_titles = pd.read_csv("Movie_Id_Titles",header=0)
movie_titles.head()
```

```python
df=pd.merge(df,movie_titles,on='item_id')
df.head(5)
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head(10)
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head(5)
```

```python
ratings['Number_of_Ratings']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating',fill_value=0)
moviemat.head()
```

```python
ratings.sort_values('Number_of_Ratings',ascending=False).head(10)
```

```python
correlation_matrix=moviemat.corr(method='pearson')
correlation_matrix.head(10)
```

```python
def get_similar(movie_name,rating):
    similar_score = correlation_matrix[movie_name]*(rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_score
```

```python
movie_lover = [("Toy Story (1995)",5),("2001: A Space Odyssey (1968)",2)]
similar_movies = pd.DataFrame()
for movie,rating in movie_lover:
    similar_movies = similar_movies.append(get_similar(movie,rating))

similar_movies.head(10)
```

```python
similar_movies.sum().sort_values(ascending=False).head(20)
```

```python
toy_story_movierating=moviemat['Toy Story (1995)']
toy_story_movierating.head(10)
```

```python
similar_to_toystory=moviemat.corrwith(toy_story_movierating)
similar_to_toystory.head(10)
```

```python
similar_df=pd.DataFrame(similar_to_toystory,columns=['Correlation'])
similar_df.head(10)
```

```python
similar_df.sort_values('Correlation',ascending=False)
```

```python
similar_df=pd.merge(similar_df,ratings,on='title')
similar_df.head(10)
```

```python
similar_df[similar_df['Number_of_Ratings']>100].sort_values('Correlation',ascending=False).head(15)
```
